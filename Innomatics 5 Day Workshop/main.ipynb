```python
x = df.drop(['MSRP'],axis=1).values
y = df['MSRP'].astype(int)
print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x,y,train_size=0.80,random_state=15) # selecting randomly 15 values
print(f'X train shape: {x_train.shape}')
print(f'X test shape: {x_test.shape}')
print(f'Y train shape: {y_train.shape}')
print(f'Y test shape: {y_test.shape}')
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression(fit_intercept=True)
```

```python
regression.fit(x_train,y_train)
```

```python
y_pred = regression.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error , root_mean_squared_error
```

```python
mse = mean_squared_error(y_test,y_pred)
```

```python
import numpy as np
print(np.sqrt(mse))
```

```python
rmse = root_mean_squared_error(y_test,y_pred)
```

```python
print(rmse)
```

```python
all_df_predict = regression.predict(x)
```

```python
mean_actual = np.mean(y_pred)
rmse_percentage = (rmse / mean_actual) * 100

print(f"Mean Actual Value: {mean_actual}")
print(f"RMSE Percentage: {rmse_percentage:.2f}%")
```
