# Data Cleaning Starts

```python
df.isnull().sum()
```

### The dataset is converted into two sets based on Region from 122th index , we can make a new column based on the Region
    0 : 'Bejaia Region Dataset'
    1 : 'Sidi-Bel Abbes Region Dataset'
### Add new column with region

```python
df.loc[:122,'Region'] = 0
df.loc[122:,'Region'] = 1
```

```python
df.info()
```

```python
df['Region'] = df['Region'].astype(int)
```

```python
df.info()
```

```python
df[df.isnull().any(axis=1)]
```

```python
df = df.dropna().reset_index(drop=True)
```

```python
df.head()
```

```python
df.iloc[[122]]
```

```python
df = df.drop(122).reset_index(drop=True)
```

```python
df.isnull().sum()
```

```python
df.columns
```

#### If you looked closely you can see that there are some spaces in the column names which we should remove

```python
df.rename(columns={'day':'Day','month':'Month'},inplace=True)
```

```python
df.columns
```

```python
df.columns = df.columns.str.strip() # Removed extra spaces
```

```python
df.columns
```

```python
df.info()
```

```python
df.columns
```

```python
df[['Day', 'Month', 'year', 'Temperature', 'RH', 'Ws']] = df[['Day', 'Month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
df.info()
```

```python
df.head()
```

```python
df.columns
```

```python
df[['Rain', 'FFMC','DMC', 'DC', 'ISI', 'BUI', 'FWI']] = df[['Rain', 'FFMC','DMC', 'DC', 'ISI', 'BUI', 'FWI']].astype(float)
```

```python
df.info()
```

```python
objects = [feature for feature in df.columns if df[feature].dtypes=='O']
for i in objects:
    if i!='Classes':
        df[i] = df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.to_csv('Algerian_forest_fires_cleaned_dataset.csv',index=False)
```
