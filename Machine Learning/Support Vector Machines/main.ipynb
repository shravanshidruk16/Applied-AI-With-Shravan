```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report , confusion_matrix
```

```python
from sklearn.datasets import make_classification
```

```python
X,y = make_classification(n_samples=1000,n_features=2,n_classes=2,n_clusters_per_class=2,n_redundant=0)
```

```python
sns.scatterplot(x=pd.DataFrame(X)[0],y=pd.DataFrame(X)[1],hue=y)
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X,y,test_size=0.25,random_state=10)
```

```python
from sklearn.svm import SVC
linear = SVC(kernel='linear')# as from scatter plot we can use linear kernel
linear.fit(X_train,y_train)
y_pred_linear = linear.predict(X_test)
print(classification_report(y_test,y_pred_linear))
print(confusion_matrix(y_test,y_pred_linear))
```

### For overlapped datasets we can use different types of kernels like - poly , rbf , sigmoid

# 1. Using RBF kernel

```python
rbf = SVC(kernel='rbf')
rbf.fit(X_train,y_train)
y_pred_rbf = rbf.predict(X_test)
print(classification_report(y_test,y_pred_rbf))
print(confusion_matrix(y_test,y_pred_rbf))
```

# 2. Using POLY (polynomial) kernel

```python
poly = SVC(kernel='poly')
poly.fit(X_train,y_train)
y_pred_poly = poly.predict(X_test)
print(classification_report(y_test,y_pred_poly))
print(confusion_matrix(y_test,y_pred_poly))
```

# 3. Using SIGMOID kernel

```python
sigmoid = SVC(kernel="sigmoid")
sigmoid.fit(X_train,y_train)
y_pred_sigmoid = sigmoid.predict(X_test)
print(classification_report(y_test,y_pred_sigmoid))
print(confusion_matrix(y_test,y_pred_sigmoid))
```

# Hyperparameter tuning using GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'C':[0.1,1,10,100,1000],
              'gamma':[1,0.1,0.01,0.001,0.0001],
              'kernel':['rbf']}
```

```python
grid = GridSearchCV(SVC(),param_grid=param_grid,cv=5,refit=True,verbose=3)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
y_pred_grid = grid.predict(X_test)
print(classification_report(y_test,y_pred_grid))
print(confusion_matrix(y_test,y_pred_grid))
```
