# Technique 1 - Removing Constant Values

```python
import numpy as np
import pandas as pd
```

```python
data = pd.DataFrame({
    "A": [1, 2, 4, 1, 2, 4],
    "B": [4, 5, 6, 7, 8, 9],
    "C": [0, 0, 0, 0, 0, 0],
    "D": [1, 1, 1, 1, 1, 1]
})
```

```python
data
```

### We will use the concept of **Variance Thresholding** where what happens is that we remove the columns according the variance threshold value if variance threshold value is 0 ie that colm will have a same value

```python
from sklearn.feature_selection import VarianceThreshold
```

```python
vt = VarianceThreshold(threshold=0)
```

```python
vt.fit(data)
```

```python
sum(vt.get_support())
```

```python
const_colms = [col for col in data.columns if col not in data.columns[vt.get_support()]]
```

```python
for useless_feature in const_colms:
    print(useless_feature)
```

```python
data = data.drop(const_colms,axis=1)
data
```
