```python
import numpy as np
import pandas as pd
from sklearn.feature_selection import VarianceThreshold
```

```python
df = pd.read_csv('train.csv',nrows=10000)
```

```python
df.shape
```

```python
df.head()
```

```python
x = df.drop(labels=['TARGET'],axis=1)
```

```python
x.head()
```

```python
y = df['TARGET']
```

```python
y.head()
```

```python
y.unique()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
x_train.shape
```

```python
vt = VarianceThreshold(threshold=0)
```

```python
vt.fit(x_train)
```

```python
total_cols = len(x_train.columns)
good_cols = sum(vt.get_support())
print(f'Good cols: {good_cols} \nBad cols: {total_cols-good_cols}')
```

```python
bad_cols = [col for col in x_train.columns if col not in x_train.columns[vt.get_support()]]
```

```python
print(len(bad_cols))
```

```python
for bad in bad_cols:
    print(bad)
```

```python
df = df.drop(bad_cols,axis=1,inplace=True)
```
