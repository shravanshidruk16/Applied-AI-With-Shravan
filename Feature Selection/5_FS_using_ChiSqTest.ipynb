```python
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
df = sns.load_dataset('titanic')
```

```python
df.shape
```

```python
df.head()
```

```python
df = df[['sex','embarked','alone','pclass','survived']]
```

```python
df.head()
```

# Before applying Chi-Square test we need to apply label encoding

```python
# sex column
df['sex'] = np.where(df['sex']=='male',1,0)
```

```python
df.head()
```

```python
# embarked column
df['embarked'].unique()
```

```python
df['embarked'] = df['embarked'].map({'S':0,'C':1,'Q':2,np.nan:3})
```

```python
df.head()
```

```python
df['embarked'].unique()
```

```python
# now alone column
df['alone'] = np.where(df['alone']==True,1,0)
```

```python
df.head()
```

# Now also a train test split in order to avoid overfitting

```python
x = df.drop('survived',axis=1)
y = df['survived']
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
x_train.head()
```

```python
from sklearn.feature_selection import chi2
```

## Note : more the value of fscore more the important the feature is , lesser the pvalue more imp the feature

```python
f_values , p_values = chi2(x_train,y_train)
```

```python
f_values
```

```python
p_values
```

```python
p_values = pd.Series(p_values)
```

```python
p_values.index=x_train.columns
```
