### Data Aggregations and Grouping 

```python
grouped_mean = df.groupby('Product_name')['Value'].mean()
print(grouped_mean)
```

```python
sales_based_on_region = df.groupby('Region')['Sales'].sum()
sales_based_on_region
```

```python
print("Aggregations: ")
grouped_agg = df.groupby('Region')['Value'].agg(['mean','sum','count'])
grouped_agg
```

### Merging and joing data frames

```python
df1 = pd.DataFrame({'Key' :['A','B','C'],'Value1':[1,2,3]})
df2 = pd.DataFrame({'Key' :['A','B','D'],'Value2':[4,5,6]})
print(df1)
print(df2)
```

```python
print("Merging dataframes based on the 'Key columns'")
pd.merge(df1,df2,on='Key',how='inner')
```

```python
pd.merge(df1,df2,on='Key',how='outer')
```
