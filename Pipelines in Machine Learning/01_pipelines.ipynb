```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler # transformation technique
from sklearn.linear_model import LogisticRegression # estimator
```

```python
steps = [('standard_scaler',StandardScaler()), # step 1 in pipeline
         ('classifier',LogisticRegression())]  # step 2 in pipeline
```

### Initializing the pipeline

```python
pipe = Pipeline(steps)
```

```python
pipe
```

```python
pipe
```

```python
from sklearn.datasets import make_classification
```

```python
X,y = make_classification(n_samples=1000)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test , y_train , y_test = train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred = pipe.predict(X_test) # Note : Here the problem of data leakage is managed by pipelines 😁
```

## Displaying pipeline or making a pipeline with : StandardScaler then Dimensionality Reduction then Estimator

```python
from sklearn.decomposition import PCA
from sklearn.svm import SVC
```

```python
steps = [('scaling',StandardScaler()), # step 1 : scaling
         ('PCA',PCA()), # step 2 : dimensionality reduction
         ('estimator',LogisticRegression())] # step 3 : estimator
```

```python
pipe2 = Pipeline(steps)
```

# You can independently apply the functions in the pipeline by using the key names 

```python
pipe2['scaling'].fit_transform(X_train)
```

```python
pipe2.fit(X_train,y_train)
```

# Complex example with column transformers

```python
import numpy as np
```

```python
from sklearn.impute import SimpleImputer
```

```python
steps = [('imputation_mean',SimpleImputer(missing_values=np.nan,strategy='mean')),
         ('scaler',StandardScaler())]
```

```python
numeric_processor = Pipeline(steps)
```

```python
numeric_processor
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
categorical_processor = Pipeline(steps=[("imputation_constant",SimpleImputer(fill_value='missing_value',strategy='constant')),
                                        ('onehot',OneHotEncoder(handle_unknown='ignore'))])
```

```python
from sklearn.compose import ColumnTransformer
```

```python
preprocessor = ColumnTransformer(
    [("categorical",categorical_processor,['gender','city']),
     ('numerical',numeric_processor,['age','height'])]
)
```

```python
preprocessor
```

```python
from sklearn.pipeline import make_pipeline
```

```python
final_pipeline = make_pipeline(preprocessor,LogisticRegression())
```
